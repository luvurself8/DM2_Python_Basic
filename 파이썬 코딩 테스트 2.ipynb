```python
seat = [ [0 for col in range(7)]for row in range(7)]

def showReservationStatus():
    print("""-----------------------
    1 2 3 4 5 6 7
-----------------------""")
    for i,r in enumerate(seat, start = 1):
        print(f"{i}|",end = "  ")
        for c in r:
            print(f"{c:2d}", end="")
        print()
                  
showReservationStatus()


def reservationSeat(row,col):
    if seat[row-1][col-1] == 1:
        return False
    else:
        seat[row-1][col-1] = 1
        return True

while True:
    choice = input("좌석을 예약하시겠습니까 (y/n):")
    if choice == "n":
        break
        
    showReservationStatus()
    howMany = int(input("몇개의 좌석을 예약하시겠습니까"))
    cnt = 0
    while cnt< howMany:
        cnt += 1
        print(f"{cnt}번째 자리 예약")
        row = int(input("행값입력:"))
        col = int(input("열값입력:"))
        if reservationSeat(row,col):
            print("예약되었습니다")
        else:
            print(f"{row}행 {col}열은 예약할 수 없습니다")
            cnt -= 1
        
```
